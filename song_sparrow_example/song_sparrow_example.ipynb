## Model Specification

```python
# creating x and y varialbles to use in pymc model
# Predictor variable
X = sparrow_data_df["age"].values
# outcome variable
y = sparrow_data_df["offspring"].values
```

```python
with pm.Model() as offspring_model:

    # define priors for unknown model parameters, weakly informative Normal
    beta = pm.Normal("beta", mu=0, sigma=10, shape=3)

    # define linear model and exp link function
    logtheta = beta[0] + beta[1] * X + beta[2] * (X*X)

    ## Define Poisson likelihood (sampling distribution) of observations
    Y_obs = pm.Poisson("Y_obs", mu=np.exp(logtheta), observed=y)
```

```python
# draw 10000 posterior samples
with offspring_model:
    idata = pm.sample(10000)
```

```python
# marginal posterior plots (smoothed histogram using kernel density estimation) and Markov chain plots
az.plot_trace(idata, combined=False, legend=True , compact=False);
```

### This is how to convert Pymc posteriors to numpy arrays

```python
posterior_array = np.asarray(idata.posterior['beta'])
np.shape(posterior_array)
```

```python
# reshape the array to combine the sample from all 4 chains in one vector
posterior_array_resh = np.reshape(posterior_array, (40000, 3))
np.shape(posterior_array_resh)
```

```python
# quick check to make sure the combining of markov chains was correct  
plt.plot(posterior_array_resh[:,2])
```

```python
# generating the quantile plots for each age(pridictor variable) using the MC sample for each unknow parameter

x_age =[1, 2, 3, 4, 5, 6]
quantiles_data = []

for x in x_age:
    # compute exp (log theta) which is equal to exp(B1 + B2*x + B3*x^2 )
    temp = np.exp(posterior_array_resh[:,0] + (posterior_array_resh[:,1]* x) +
              (posterior_array_resh[:,2]* x**2))
    
    quantiles= np.quantile(temp, (0.025, 0.5, 0.975),  axis=0)
    
    quantiles_data.append(
    {
        'age'             : x,
        'offspring_2.5qt' : quantiles[0],
        'offspring_50qt'  : quantiles[1],
        'offspring_97.5qt': quantiles[2],
        
    }) 
    
```

```python
quantiles_df = pd.DataFrame(quantiles_data)
quantiles_df
```

```python
# plot the quantiles
quantiles_df.plot(x = 'age')
plt.ylabel('nmuber of offspring')
```
